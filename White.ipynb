```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
data = pd.read_csv("Death_rates_for_suicide__by_sex__race__Hispanic_origin__and_age__United_States.csv")
```

```python
# Filter the dataset to include males who are White or African American between the ages of 25-44 years and the years 1950 and 1984
White_Male_Data = data[(data['STUB_LABEL'].str.contains('Male: White')) &  
                     (data['YEAR'].between(1985, 2018))]

# Display the filtered dataset including the INDICATOR, UNIT, AGE, YEAR, and ESTIMATE columns
White_Male_Data[['INDICATOR', 'UNIT', 'STUB_LABEL', 'AGE', 'YEAR', 'ESTIMATE']]
```

```python
White_Female_Data = data[(data['STUB_LABEL'].str.contains('Female: White')) & 
                     (data['YEAR'].between(1985, 2018))]

White_Female_Data[['INDICATOR', 'UNIT', 'STUB_LABEL', 'AGE', 'YEAR', 'ESTIMATE']]
```

```python
# Create a single plot
plt.figure(figsize=(12, 8))

# Define colors for each age category
colors = ['blue', 'green', 'red', 'orange', 'purple', 'cyan', 'magenta', 'yellow', 'lime', 'pink']

# Plot line plots for each age category for White Males
for i, age_category in enumerate(['15-24 years', '25-44 years', '45-64 years', '65-74 years', '75-84 years']):
    # Filter data for White Males of the current age category
    male_data = White_Male_Data[White_Male_Data['AGE'] == age_category]
    
    # Plot line plot for White Males of the current age category with a different color
    plt.plot(male_data['YEAR'], male_data['ESTIMATE'], color=colors[i], marker='o', label=f'White Males ({age_category})')

# Plot line plots for each age category for White Females
for i, age_category in enumerate(['15-24 years', '25-44 years', '45-64 years', '65-74 years', '75-84 years']):
    # Filter data for White Females of the current age category
    female_data = White_Female_Data[White_Female_Data['AGE'] == age_category]
    
    # Plot line plot for White Females of the current age category with a different color
    plt.plot(female_data['YEAR'], female_data['ESTIMATE'], color=colors[i+len(colors)//2], marker='o', linestyle='--', label=f'White Females ({age_category})')

# Set titles and labels
plt.title('Suicide Estimates Over Time for White Males and Females by Age Category')
plt.xlabel('Year')
plt.ylabel('Suicide Estimate (% per 100,000 resident population)')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
